# Functions

```python
#function to extrat 

#title
def get_title(soup):
    try:
        title_string = soup.find('div', class_='projectHeading').get_text().strip()
    except (AttributeError,IndexError):
        title_string=""
        
    return title_string

#price
def get_price(soup):
    try:
        price= soup.find('div', class_='withDeposite').get_text().strip().split()[1]
    except (AttributeError,IndexError):
        price=""
        
    return price

#deposit
def get_deposit(soup):
    try:
        deposit = soup.find('div', class_='withDeposite').get_text().strip().split()[4]
    except (AttributeError,IndexError):
        deposit=""
        
    return deposit
#listing Type
def get_listing_type(soup):
    try:
        listing_Type = soup.find_all('div',class_='projectLiData')[0].get_text().replace('Listing Type',"")
    except (AttributeError,IndexError):
        listing_Type=""
        
    return listing_Type

#Property Type
def get_Property_Type(soup):
    try:
        Property_Type = soup.find('div', class_='projectHeading').get_text().strip()
    except (AttributeError,IndexError):
        Property_Type=""
        
    return Property_Type

#Locality
def get_Locality(soup):
    try:
        Locality = soup.find_all('div',class_='projectLiData')[4].get_text().replace('Locality',"")
    except (AttributeError,IndexError):
        Locality=""
        
    return Locality

#area
def get_area(soup):
    try:
        area = soup.find_all('div',class_='projectLiData')[5].get_text().replace('Area',"")
    except (AttributeError,IndexError):
        area=""
        
    return area
#Furnishing_Status
def get_Furnishing_Status(soup):
    try:
        Furnishing_Status = soup.find_all('div',class_='projectLiData')[7].get_text().replace('Furnishing Status ',"")
    except (AttributeError,IndexError):
        Furnishing_Status=""
        
    return Furnishing_Status

#Number_of_Room
def get_Number_of_Room(soup):
    try:
        Number_of_Room = soup.find_all('div',class_='projectLiData')[8].get_text().replace('Number of Rooms',"")
    except (AttributeError,IndexError):
        Number_of_Room=""
        
    return Number_of_Room
#Number_of_Bathroom
def get_Number_of_Bathroom(soup):
    try:
        Number_of_Bathroom = soup.find_all('div',class_='projectLiData')[9].get_text().replace('Number of Bathroom',"").strip()
    except (AttributeError,IndexError):
        Number_of_Bathroom=""
        
    return Number_of_Bathroom

#Number_of_Parking
def get_Number_of_Parking(soup):
    try:
        Number_of_Parking = soup.find_all('div',class_='projectLiData')[10].get_text().replace('Number of Parking',"").strip()
    except (AttributeError,IndexError):
        Number_of_Parking=""
        
    return Number_of_Parking
#Floor_No
def get_Floor_No(soup):
    try:
        Floor_No = soup.find_all('div',class_='projectLiData')[12].get_text().replace('Floor No.',"").strip()
    except (AttributeError,IndexError):
        Floor_No=""
        
    return Floor_No
#Security_Deposit
def get_Security_Deposit(soup):
    try:
        Security_Deposit = soup.find_all('div',class_='projectLiData')[13].get_text().replace('Security Deposit',"").strip()
    except (AttributeError,IndexError):
        Security_Deposit=""
        
    return Security_Deposit
```

```python
if __name__=='__main__':
    #headers for request
    Headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",'Accept-Language':'en-US, en;q=0.5' }
    
    #the URL
    URL ='https://www.squareyards.com/rent/property-for-rent-in-nashik'
    
    #HTTP for request
    webpage_new = requests.get(URL,headers=Headers)
    
    #soup object contaning all the data
    soup = BeautifulSoup(webpage_new.content,'html.parser')
    
    #fetch the links as list of tags object 
    
    link = soup.find_all('button', class_='tlSrtlst shortlistcontainerlink1 DSE_Rental_D19')
    
    #lopp for extracting the product links
    LINK=[]
    for i in range(len(link)):
        a=link[i].get('data-propertylink')
        LINK.append(a)
    d={'Title':[],'Price':[],'Deposit':[],'Listing_Type':[],'Property_Type':[],'Locality':[],'Area':[],'Furnishing_Status':[],
      'Number_of_Room':[],'Number_of_Bathroom':[],'Number_of_Parking':[],'Floor_No':[],'Security_Deposit':[]}
    
    #Loop for extracting the product details from each link
    
    for link in LINK:
        new_webpage= requests.get(link,headers=Headers)
        new_soup= BeautifulSoup(new_webpage.content,'html.parser')
        
        #functions to call to display all necessary info
        
        d['Title'].append(get_title(new_soup))
        d['Price'].append(get_price(new_soup))
        d['Deposit'].append(get_deposit(new_soup))
        d['Listing_Type'].append(get_listing_type(new_soup))
        d['Area'].append(get_area(new_soup))
        d['Property_Type'].append(get_Property_Type(new_soup))
        d['Locality'].append(get_Locality(new_soup))
        d['Furnishing_Status'].append(get_Furnishing_Status(new_soup))
        d['Number_of_Room'].append(get_Number_of_Room(new_soup))
        d['Number_of_Bathroom'].append(get_Number_of_Bathroom(new_soup))
        d['Number_of_Parking'].append(get_Number_of_Parking(new_soup))
        d['Floor_No'].append(get_Floor_No(new_soup))
        d['Security_Deposit'].append(get_Security_Deposit(new_soup))

    
```
