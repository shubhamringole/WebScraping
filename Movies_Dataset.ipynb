```python
#installing liibraries
from bs4 import BeautifulSoup, SoupStrainer
import requests
import datetime
import pandas as pd
import numpy as np
from lxml import etree
import time
```

```python
#Defining function to extract data
#moviename
def get_movie_name(soup):
    try:
        name=soup.find('div', class_='title ott_false').get_text().strip().split('\n')[0]
        
    except (AttributeError,IndexError):
        try:
            name = soup.find('div', class_='title ott_true').get_text().strip().split('\n')[0]
        except (AttributeError,IndexError):
            name = np.nan

    return name

#certification
def get_certification(soup):
    try:
        cert = soup.find('span', class_='certification').get_text().strip()
    except (AttributeError,IndexError):
        cert = np.nan
    return cert
#release date
def get_release_date(soup):
    try:
        date = soup.find('span', class_='release').get_text().strip().split()[0]
    except (AttributeError,IndexError):
        date = np.nan
    return date
#genres
def get_genres(soup):
    try:
        genres = soup.find('span', class_='genres').get_text().strip().split('\xa0')
    except (AttributeError,IndexError):
        genres=np.nan
    return genres

#language
def get_language(soup):
    try:
        language= soup.find('section', class_='facts left_column').get_text().strip().split()[4]
    except (AttributeError,IndexError):
        language= np.nan
    return language

#budget
def get_budget(soup):
    try:
        language= soup.find('section', class_='facts left_column').get_text().strip().split()[6]
    except (AttributeError,IndexError):
        language= np.nan
    return language

#revenue
def get_revenue(soup):
    try:
        language= soup.find('section', class_='facts left_column').get_text().strip().split()[-1]
    except (AttributeError,IndexError):
        language= np.nan
    return language

#runtime
def get_runtime(soup):
    def hrs_to_min(a):
            a= a.split()
            b = int(a[0])*60 + int(a[1])
            return b
    try:
        a= soup.find('span', class_='runtime').get_text().strip().replace("h","").replace('m',"")
        time = hrs_to_min(a)
    except (AttributeError,IndexError):
        time = np.nan
    return time
        
```

```python
if __name__ == '__main__':
    d = {'Movie_Name':[],'Certification':[],'Release_Date':[],'Genres':[],'Language':[],'Budget':[],'Revenue':[],'Runtime':[]}
    #lets get all the links
    LINK=[]

    for i in range(1,200): # if u need to scrap onply 5 pages put 5 +1 insted of 100
        LINK.append('https://www.themoviedb.org/movie?page='+str(i))
    
    #page wise 
    for link in LINK:
        URL= link
        #Headers for request
        Headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",'Accept-Language':'en-US, en;q=0.5' }
        new_webpage= requests.get(URL,headers=Headers)
        new_soup= BeautifulSoup(new_webpage.content,'lxml')
        print(link)
        SUB_LINK=[]
        def sub_links(soup):
            sub_link = soup.find_all('a', class_='image')
            for i in range(len(sub_link)):
                num= sub_link[i].get('href')
                name=sub_link[i].get('title').strip().replace(" ",'-')
                link='https://www.themoviedb.org/'+str(num)+'-'+ name
                SUB_LINK.append(link)
        #getting all conecnts for page
        sub_links(new_soup)
        for sub_link in SUB_LINK:
            URL_a= sub_link
            #Headers for request
            Headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",'Accept-Language':'en-US, en;q=0.5' }
            new_webpage_a= requests.get(URL_a,headers=Headers)
            new_soup_a= BeautifulSoup(new_webpage_a.content,'html.parser')
            
            #appedning the data into our dictionery 
            
            d['Movie_Name'].append(get_movie_name(new_soup_a))
            d['Certification'].append(get_certification(new_soup_a))
            d['Release_Date'].append(get_release_date(new_soup_a))
            d['Genres'].append(get_genres(new_soup_a))
            d['Language'].append(get_language(new_soup_a))
            d['Budget'].append(get_budget(new_soup_a))
            d['Revenue'].append(get_revenue(new_soup_a))
            d['Runtime'].append(get_runtime(new_soup_a))
            
```

```python
df = pd.DataFrame.from_dict(d)
```
